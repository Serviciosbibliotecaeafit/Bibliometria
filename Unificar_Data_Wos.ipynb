###Librerias

```python
import pandas as pd
```

###Cargar la data

```python
outPath = 'output/'
filePath = 'WOS/'
```

```python
df1= pd.read_csv(filePath+"AL_1.txt",sep="\t")
df2= pd.read_csv(filePath+"AL_2.txt",sep="\t")
df3= pd.read_csv(filePath+"AL_3.txt",sep="\t")
df4= pd.read_csv(filePath+"BR_1.txt",sep="\t")
df5= pd.read_csv(filePath+"BR_2.1.txt",sep="\t")
df6= pd.read_csv(filePath+"BR_2.2.txt",sep="\t")
```

###Dimensiones

```python
df1.columns
```

```python
print(df1.shape, df2.shape, df3.shape, df4.shape, df5.shape, df6.shape)
```

###Organización

```python
unificado_AL = pd.concat ([df1,df2,df3], axis=0)
unificado_BR = pd.concat ([df4,df5,df6], axis=0)
```

```python
print(unificado_AL.shape, unificado_BR.shape)
```

```python
unificado_AL = unificado_AL.drop_duplicates(subset= ['TI', 'AF'], keep='last')
unificado_AL.shape
```

```python
unificado_BR = unificado_BR.drop_duplicates(subset= ['TI', 'AF'], keep='last')
unificado_BR.shape
```

###Exportación

```python
unificado_AL.to_csv(outPath+'WOS_AL.csv')
unificado_BR.to_csv(outPath+'WOS_BR.csv')
```

```python
Exportación = pd.ExcelWriter("Data_WOS.xlsx")
unificado_AL.to_excel(Exportación, sheet_name="WOS_AL")  
unificado_BR.to_excel(Exportación, sheet_name="WOS_BR")
Exportación.save()
```
