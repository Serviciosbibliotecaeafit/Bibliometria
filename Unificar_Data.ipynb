# Unificación de Bases de Datos
Bases de datos a unificar: Lens, Scielo, Scopus, WoS. Cada una con variación Latinoamerica(AL) y Brasil(BR).

```python
import pandas as pd # Libreria para usar el objeto dataFrame
from utilities import * # Archivo con diccionarios
import numpy as np
# Paths
dataPath = 'output/'
dataBases = ['LENS', 'SCIELO', 'SCOPUS', 'WOS']
# import data
data_AL = []
data_BR = []
for base in dataBases:
    data_AL.append(pd.read_csv(dataPath+base+'_AL.csv',sep=','))
    data_BR.append(pd.read_csv(dataPath+base+'_BR.csv',sep=','))

dictionaries_en = [Lens_Dictionary(lang='en'),
                   Scielo_Dictionary(),
                   Scopus_Dictionary(),
                   WoS_Dictionary()]
dictionaries_es = [Lens_Dictionary(lang='es'),
                   Scielo_Dictionary(),
                   Scopus_Dictionary(),
                   WoS_Dictionary()]

unified_data_AL = {}
unified_data_BR = {}
unified_data_AL = full_Process(data_AL,dataBases, dictionaries_en)
unified_data_BR = full_Process(data_BR,dataBases, dictionaries_es)
```
