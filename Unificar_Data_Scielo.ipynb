###Librerias

```python
import pandas as pd
```

###Cargar la data

```python
outPath = 'output/'
filePath = 'SCIELO/'
```

```python
df1= pd.read_csv(filePath+"AL_1.csv",sep=",")
df2= pd.read_csv(filePath+"AL_2.csv",sep=",")
df3= pd.read_csv(filePath+"BR_1.csv",sep=",")
df4= pd.read_csv(filePath+"BR_2.csv",sep=",")
```

###Dimensiones

```python
df1.columns
```

```python
print(df1.shape, df2.shape, df3.shape, df4.shape)
```

###Organización

```python
unificado_AL = pd.concat ([df1,df2], axis=0)
unificado_BR = pd.concat ([df3,df4], axis=0)
```

```python
unificado_AL.shape
```

```python
unificado_BR.shape
```

```python
unificado_AL = unificado_AL.drop_duplicates(subset= ['Title', 'Author(s)'], keep='last')
unificado_AL.shape
```

```python
unificado_BR = unificado_BR.drop_duplicates(subset= ['Title', 'Author(s)'], keep='last')
unificado_BR.shape
```

###Exportación

```python
unificado_AL.to_csv(outPath+'SCIELO_AL.csv')
unificado_BR.to_csv(outPath+'SCIELO_BR.csv')
```

```python
Exportación = pd.ExcelWriter("Data_scielo.xlsx")
unificado_AL.to_excel(Exportación, sheet_name="Scielo_AL")  
unificado_BR.to_excel(Exportación, sheet_name="Scielo_BR")
Exportación.save()
```
