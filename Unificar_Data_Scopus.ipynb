###Librerias

```python
import pandas as pd
```

###Cargar la data

```python
outPath = 'output/'
filePath = 'SCOPUS/'
```

```python
df1 = pd.read_csv(filePath+"AL_1.csv", sep=",")
df2 = pd.read_csv(filePath+"AL_2.csv", sep= ",")
df3 = pd.read_csv(filePath+"AL_3.csv", sep= ",")
df4 = pd.read_csv(filePath+"BR_1.csv", sep= ",")
df5 = pd.read_csv(filePath+"BR_2.csv", sep= ",")
```

###Dimensiones

```python
df1.columns
```

```python
print(df1.shape, df2.shape, df3.shape, df4.shape, df5.shape)
```

###Organización

```python
unificado_AL = pd.concat ([df1,df2,df3], axis=0)
unificado_BR = pd.concat ([df4,df5], axis=0)
```

```python
print(unificado_AL.shape, unificado_BR.shape)
```

```python
unificado_AL = unificado_AL.drop_duplicates(subset= ['Title', 'Authors'], keep='last')
unificado_AL.shape
```

```python
unificado_BR = unificado_BR.drop_duplicates(subset= ['Title', 'Authors'], keep='last')
unificado_BR.shape
```

###Exportación

```python
unificado_AL.to_csv(outPath+'SCOPUS_AL.csv')
unificado_BR.to_csv(outPath+'SCOPUS_BR.csv')
```

```python
Exportación = pd.ExcelWriter("Data_scopus.xlsx")
unificado_AL.to_excel(Exportación, sheet_name="Scopus_AL")  
unificado_BR.to_excel(Exportación, sheet_name="Scopus_BR")
Exportación.save()
```
